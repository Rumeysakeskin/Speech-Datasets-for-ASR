```python
import librosa
from os import listdir
from os.path import isfile, join
import json
```

# Load wav and txt files

```python
text_path = "./Voxforge/voxforge_dataset/txt"
txtfiles = [f for f in listdir(text_path) if isfile(join(text_path, f))]
audio_path = "./Voxforge/voxforge_dataset/wav"
wavfiles = [f for f in listdir(audio_path) if isfile(join(audio_path, f))]
```

```python
len(wavfiles), len(txtfiles)
```

# Create manifest.jsonl

```python
output_file = "voxforge_tr_dataset_manifest.jsonl"
total = 0
with open(output_file, 'w', encoding="utf-8") as file:

        
    for txtfile in txtfiles:
        if txtfile.split(".txt", 1)[0] + ".wav"  in wavfiles:
            
            wavfile = txtfile.split(".txt", 1)[0] + ".wav"
            
            wav_path = audio_path + "/" + wavfile
            duration = librosa.core.get_duration(filename=wav_path)

            textfile_path = text_path + "/" + txtfile
            with open(textfile_path, 'r', encoding="utf-8") as f:  
                for text_cap in f:
                    text = text_cap.lower()

            file.write(json.dumps({
                'audio_filepath': wav_path,
                "duration": duration,
                'text': text},  ensure_ascii=False) + '\n')
            
        else:
            print("Some files not processed!")
            
        total+= 1
print("{} files processed".format(total))    
print("done!")
```

# Play audios

```python
import IPython.display as ipd

from nemo.collections.asr.parts import perturb, segment
# NeMo has its own support class for loading wav files
def load_audio(filepath) -> segment.AudioSegment:
    sample_segment = segment.AudioSegment.from_file(filepath, target_sr=16000)
    return sample_segment
```

```python
filepath = "./Voxforge/voxforge_dataset/wav/5Forward-20140616-bpk_tr-0007.wav"
sample_segment = load_audio(filepath)
ipd.Audio(sample_segment.samples, rate=16000)
```

```python
import json
import os
voxforge_duration = 0
manifest_path = 'voxforge_tr_dataset_manifest.jsonl'
with open(manifest_path, 'r', encoding="utf-8") as f:  
    for line in f:
        original_file_json = json.loads(line)
        voxforge_duration += original_file_json["duration"]

print("TR Voxforge hour:",voxforge_duration/3600)
```
